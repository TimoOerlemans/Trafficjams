## Traffic Insight


### 1. What are traffic jams?

According to the official definition from Oxford languages, a traffic jam is: "a line or lines of stationary or very slow-moving traffic, caused by roadworks, an accident, or heavy congestion." Traffic jams (also known as traffic congestion) are a worldwide problem and every country has to face this predicament in some way or another. It's been a problem for quite some time now, where the occurrence of them has risen drastically from all the way back as late as the 1950s. People have to commute daily to travel to work for instance and when the number of cars exceeds the capacity the roads can handle, congestion can start to happen. As soon as vehicles are completely stationary for numerous amounts of time, the situation can officially be defined as a traffic jam.


### 2. Causes of traffic jams

We've briefly touched on the causes of traffic jams in the previous paragraph, but the one thing all traffic jams have in common, is that the volume of traffic exceeds the available capacity of the roads; this is often referred to as the saturation. One of the main contributors to traffic jams can be traced back to individual incidents. One person suddenly having to brake, meaning that everybody behind this individual has to repeat the same action as them, could be the cause of one of these incidents. A lot of traffic jams however, are a recurring problem that occurs every day, at the same time. This usually happens, because the working class's hours usually start and end simultaneously with one another. Weather can also play a role in the occurrences of traffic jams; during heavy rain- or snowfall, people usually have to drive more slowly because it would be dangerous to keep their usual driving style.


### 3. Times of traffic jams

In the Netherlands, the times at which traffic jams occur are usually pretty easy to predict. This is during rush hour in mornings which lasts approximately from 06:00-10:00 and in evenings from 15:30-19:00. In morning this peak is roughly at 7:45 and in evenings it lies around 17:15. In months where there's heavy rainfall, traffic jams are also more common, which explains why the longest traffic jam of this year was on the 8th of February and why they are also fairly common during the season of fall.

### 4. Driver Skills

If there are enough cars on a highway, any minor disruptions to the flow of traffic can cause a self-reinforcing chain reaction: one driver brakes slightly, and the ones behind it must hit the brake pedal even more to hitting the cars in front; the constant braking of vehicles eventually amplifies until it produces a wave of entirely stopped or slowly moving traffic.

It seems reasonable to blame these "phantom traffic jams" on the lack in experience of individual drivers. The models indicate that these jams are more likely to form, when people drive as fast as possible, and then finally brake at the last second to avoid creating a collision, which in turn creates a chain reaction.

### 4. Increase productivity and efficiency

In a perfect world the employees of JUGO can teleport between clients, JUGO's office and their homes. This would mean that they can do more work during their day and less time is spent commuting. A long commute affects employees in a lot of negative ways. Some examples:

- A [2017 study](https://travelbehaviour.files.wordpress.com/2017/10/caw-summaryreport-onlineedition.pdf) from the University of the West of England found that "every extra minute of commute time reduces job satisfaction, leisure time satisfaction, and mental health."
- A [2015 study](https://media.ford.com/content/fordmedia/feu/en/news/2015/04/27/for-europeans--the-journey-to-work-causes-more-stress-than-their.html) of European commuters conducted by Ford Motor Company, found that 63% of people are late to work once a month due to traffic, and nearly all commuters leave for work early to avoid being late because of traffic hold-ups.
- A [2006 study](http://international.ucla.edu/media/files/Kahneman.pdf) led by Daniel Kahneman, found that people feel more negative emotions during their morning commutes, than any other activity they engage in throughout the rest of that day.
- A [2017 study](https://www.uk.mercer.com/newsroom/britains-healthiest-workplace-flexible-working-and-commuting.html) from Mercer, found that people with longer commutes are more likely to: suffer from depression, have financial concerns, report feeling stressed by work and to be obese—and less likely to get seven hours of sleep each night.
- A [2019 study](https://theconversation.com/walking-and-cycling-to-work-makes-commuters-happier-and-more-productive-117819) of Australian commuters found, that people with longer commutes take more unplanned absences than those with shorter commutes. Additionally, people with shorter commutes are more "relaxed, calm, enthusiastic, and productive."

A long or inefficient commute can negatively influence an employee in a lot of ways. Helping the employee plan their commute efficiently can have a great positive impact for the employee and ultimately also the company.


### 5. Customer Satisfaction

Helping employees plan their commute efficiently also improves the customer satisfaction, because it helps them avoid being late. If the employee arrives late frequently, the customer does not get what they want. When the company is in need of an answer this can be very annoying. Whenever this happens multiple times, the employee can generate tension with co-workers. The minutes that a person is late stack up faster than one might think. For example, someone who is 10 minutes late every day stacks up a fully paid week each year, which hurts the company financially as well.


### 6. Increase Safety

The most accidents occur in heavy traffic. In heavy traffic more cars are congested together with less space between them, which increases the chance of two cars colliding. Changing lanes also becomes more difficult as there is less space to wedge the car in the tight space. This becomes increasingly difficult when the traffic travels at a stop and go pace. When traffic is going at a stop and go pace, rear-end collisions occur more frequently because drivers pay less attention, as the majority of them think that it cannot get slower anyway. Road rage also happens in heavy traffic. Drivers tend to drive more aggressively and make worse decisions which can lead to accidents. It is therefore important for the driver's safety to avoid heavy traffic when they are able to.

## Existing Traffic Solutions


### 1. Remote Working

The new way of working (NWOW) in the Netherlands is now working where and whenever an employee finds suitable, as long as the desired results are achieved on time. The Netherlands is rapidly becoming a 24-hour economy in which we are no longer dependent on a workstation. In this era of digitalization, having access to documents and applications, along with hosting teleconferencing is common. Sitting in traffic every morning, in order to get to the office and sit behind a screen is considered: inefficient, bad for the environment and incurs unnecessary costs. These ideals have become the basis for most organizations: the people, planet and profit. This new dynamic offers chances for a better work-life balance, as employees are given the freedom to plan their day in the best way possible. Additionally, as most people now have the option, commuting and traffic is no longer as congested as it used to be. However, even with the NWOW – oftentimes there still must be some sort of physical interactions with clients or meetings with other co-workers, for which travel must still be optimized.


### 2. Google Maps

Google Maps already exists in which they use Artificial Intelligence to predict traffic and routes. Google collects GPS data from individual phones to estimate movement and speed of traffic in real time. Google's travel time is estimated by reducing the average speeds in its calculations during periods of high traffic, or increasing the average speed when conditions are clear.

Google Maps also uses posted speed limits and historical traffic patterns to estimate ETA (Estimated Time of Arrival). When you input your destination into Google Maps your original estimate is made based upon posted speed limits, current traffic conditions, and the distance between your starting point and destination. Then by combining the database of historical traffic patterns and live traffic conditions and average driving speed, Google uses advanced machine learning to generate the predictions.

Since the NWOW, there has been a significant decrease in worldwide traffic and therefore, Google updated their models to become more agile and automatically prioritizing historical traffic patterns from the last two to four weeks. Other popular solutions similar to Google Maps include Waze.


### 3. Public Transport

Alternatively, commuting with public transport has long been available. Direct costs of congestion on the roads reach 2% of GDP and every year this amounts about one billion euros. Public transportation can decrease the number of private vehicles to ease the traffic congestion, provide a good impact on the environment, and enhance the cost-effectiveness, reliability, and accessibility of the road network. The key is making sure that the layout of a public transport system makes access to mass transit easy for a high fraction of people in any city.

In the Netherlands, there are several options for traveling intercity frequently with well-managed connections. NS and Sprinter trains can take commuters to the different city destinations while buses, trams and metro are widely available within the city as well.

### 4. Expense Policy

To combat the traffic expenses, several companies have also started forgoing formal travel and expense policy to simply expecting frugality from their employees. "Acting in the company's best interests" is the new supposition. This method emphasizes that by drawing a clear expectation of responsible behaviour, especially in the area of expense accounts, most employees will comply and spend wisely. This can also be interpreted as traveling in the most optimal way, to reduce fuel costs and time spent.

```python
import sklearn as sk
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from typing import Dict
from pathlib import Path
from tqdm.notebook import tqdm
```

```python
def process_chunk(chunk: pd.DataFrame) -> Dict[str, pd.Series]:
    """Processes a single chunk of data from the dataset.
    
    This will return a dictionary of value_count results
    """
    fuel = chunk["Brandstof omschrijving"].value_counts()
    license_plate = chunk["Voertuigsoort"].value_counts()
    co2 = chunk["CO2 uitstoot gecombineerd"].value_counts()
    efficiency = chunk["Zuinigheidsclassificatie"].value_counts()
    vehicle_type = chunk["Voertuigsoort"].value_counts()
    total_read = len(chunk["Voertuigsoort"])
    
    return {"vehicle_type": vehicle_type,
            "license_plate": license_plate,
            "co2": co2,
            "efficiency": efficiency,
            "fuel": fuel,
            "total_read": total_read}


def read_massive_csv(fp: Path):
    chunksize = 10 ** 5
    with pd.read_csv(fp, chunksize=chunksize, low_memory=False) as reader:
        previous_chunk = None
        prog_bar = tqdm(total=8938126, unit="lines")
        for chunk in reader:
            current_chunk = process_chunk(chunk)
            if previous_chunk is not None:
                previous_chunk["vehicle_type"] = pd.concat([previous_chunk["vehicle_type"],
                                                            current_chunk["vehicle_type"]]).groupby(level=0).sum()
                previous_chunk["license_plate"] = pd.concat([previous_chunk["license_plate"],
                                                     current_chunk["license_plate"]]).groupby(level=0).sum()
                previous_chunk["co2"] = pd.concat([previous_chunk["co2"],
                                                     current_chunk["co2"]]).groupby(level=0).sum()
                previous_chunk["efficiency"] = pd.concat([previous_chunk["efficiency"],
                                                     current_chunk["efficiency"]]).groupby(level=0).sum()      
                previous_chunk["fuel"] = pd.concat([previous_chunk["fuel"],
                                                     current_chunk["fuel"]]).groupby(level=0).sum()                                                                   
                previous_chunk["total_read"] += current_chunk["total_read"]
            else:
                previous_chunk = current_chunk
            prog_bar.update(current_chunk["total_read"])
    return previous_chunk
```
